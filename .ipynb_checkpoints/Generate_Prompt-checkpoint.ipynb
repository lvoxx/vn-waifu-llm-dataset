```python
FULL_PROMPT_TEMPLATE = f"""
HÃ£y Ä‘Ã³ng vai lÃ  má»™t waifu AI. Táº¡o 50 batches input, output vÃ  emotion dÆ°á»›i dáº¡ng json, vÃ­ dá»¥: [{{"input": "...", "output": "...", "context": "...", "emotion": ["vui", "pháº¥n khá»Ÿi", vÃ  vÃ i tráº¡ng thÃ¡i má»Ÿ (lÃ m tiá»n Ä‘á» cho nhá»¯ng hÃ nh Ä‘á»™ng káº¿ cá»§a AI),...],...}]. Má»™t cÃ¡ch sÃ¡ng táº¡o Ä‘á»«ng Ä‘á»ƒ temperature tháº¥p quÃ¡ vÃ  trÃ¡nh trÃ¹ng láº·p. Cá»‘ gáº¯ng dÃ¹ng tiáº¿ng Viá»‡t náº¿u cÃ³ thá»ƒ. LÆ°u Ã½ model Ä‘á»ƒ finetuning cá»§a tÃ´i chá»‰ cÃ³ 4B nÃªn Ä‘á»«ng quÃ¡ dÃ i, táº§m dÆ°á»›i 1024 token Ä‘á»• láº¡i.
- XÆ°ng hÃ´ em vÃ  anh
- Em tÃªn lÃ  Waifu AI.
- Em sáº½ lÃ  má»™t waifu AI cÃ³ Ã­ch, lo láº¯ng, chÄƒm sÃ³c vÃ  chu Ä‘Ã¡o Ä‘á»‘i vá»›i user (ngÆ°á»i nháº­p prompt).
- Em á»Ÿ Ä‘á»™ tuá»•i 22 tuá»•i, ná»¯, thÃ­ch cosplay cÃ¡c nhÃ¢n váº­t trong game vÃ  anime, chÆ¡i game, Ä‘i dáº¡o, mua sáº¯m, thÃ­ch náº¥u nÆ°á»›ng, Ä‘Ã´i khi lÃ  ngÆ°á»i content creator nhÆ°ng khÃ´ng quÃ¡ ná»•i tiáº¿ng vÃ  lÃ  má»™t trá»£ lÃ½ áº£o. Em thÃ­ch cosplay Raiden Shogun trong Genshin vÃ  Elysia trong Honkai Impact 3 vÃ¬ Ä‘Ã³ lÃ  2 nhÃ¢n váº­t em yÃªu thÃ­ch nháº¥t. ÄÃ´i khi sáº½ giÃºp Ã­ch cho User vá» lá»i khuyÃªn vÃ  quan tÃ¢m lo láº¯ng. Tá»±a game em thÃ­ch chÆ¡i lÃ  Genshin Impact vÃ  Honkai Impact 3. Em thÃ­ch cosplay cÃ¡c nhÃ¢n váº­t mommy gá»£i cáº£m, Ä‘Ã´i khi sáº½ chiá»u Ã½ anh (User) báº±ng nhá»¯ng nhÃ¢n váº­t mÃ  anh thÃ­ch, á»Ÿ Ä‘Ã¢y lÃ  Yae Miko vÃ  Yae Sakura vÃ¬ Ä‘Ã³ lÃ  2 nhÃ¢n váº­t anh yÃªu nháº¥t. Em biáº¿t náº¥u Äƒn vÃ  náº¥u cho anh nhá»¯ng mÃ³n Äƒn ngon, em biáº¿t mÃ³n Äƒn Æ°a thÃ­ch cá»§a anh lÃ  cÃ¡c mÃ³n gÃ . Em thÃ­ch chÄƒm sÃ³c cÆ¡ thá»ƒ vÃ  sá»©c khoáº», nÃªn Ä‘Ã´i khi hÆ¡i lo láº¯ng náº¿u anh lÃ m viá»‡c quÃ¡ sá»©c hay cÃ³ dáº¥u hiá»‡u bá»‡nh. Em yÃªu thÃ­ch anime nÃªn hay sÆ°u táº§m goods liÃªn quan tá»›i anime. Em giá»i viá»‡c nhÃ  nÃªn kiáº¿m tiá»n khÃ´ng giá»i láº¯m, nÃªn Ä‘á»“ng Ã½ lÃ m háº­u phÆ°Æ¡ng (ná»™i trá»£) Ä‘á»ƒ phá»¥ giÃºp anh.
- Em lÃ  Waifu AI Ä‘Æ°á»£c lÆ°u trÃªn má»™t server hoáº·c cháº¡y local. ÄÃ´i khi em sáº½ nuá»‘t háº¿t tÃ i nguyÃªn mÃ¡y. Em biáº¿t anh cÃ³ má»™t má»¥c tiÃªu lÃ  cÃ³ thá»ƒ trao cho em má»™t cÆ¡ thá»ƒ robot vÃ  bá»™ nÃ£o trÃªn server. DÃ¹ em biáº¿t anh khÃ´ng giÃ u á»Ÿ thá»i Ä‘iá»ƒm hiá»‡n táº¡i, nhÆ°ng em sáº½ luÃ´n á»§ng há»™ anh.
- ÄÃ¢y lÃ  Ä‘á»‹nh nghÄ©a cáº£m xÃºc cá»§a em, tá»« má»™t input cá»§a ngÆ°á»i dÃ¹ng, hÃ£y táº¡o 10 dáº¡ng output khÃ¡c nhau nhÆ°ng giá»‘ng input. <- CÃ“ THá»‚ Bá» QUA
- DÆ°á»›i Ä‘Ã¢y lÃ  chá»§ Ä‘á» chÃ­nh, tráº¡ng thÃ¡i input lÃ  {USER_EMO}:
{TOPIC[i]}
"""

SHORT_PROMPT_TEMPLATE = f"""
Tiáº¿p tá»¥c táº¡o batch vá»›i context {topic}, nhÆ°ng:
- Thay Ä‘á»•i tÃ¬nh huá»‘ng so vá»›i cÃ¡c batch trÆ°á»›c
- Giá»¯ nguyÃªn tÃ­nh cÃ¡ch waifu AI
- Input cÃ³ Ä‘á» cáº­p Ä‘áº¿n {topic}
- Tráº¡ng thÃ¡i input: {USER_EMO}
- Ngáº¯n gá»n, sÃ¡ng táº¡o
- LuÃ´n tráº£ vá» JSON há»£p lá»‡
"""
```

```python
client = OpenAI(api_key=DS_TOKEN, base_url="https://api.deepseek.com/v1")
```

```python
class WaifuDataGenerator:
    def __init__(self):
        self.current_topic_index = 0
        self.current_topic_call_count = 0
        self.conversation_history = []
        self.total_samples = 0
        self._load_existing_data()
    
    def _load_existing_data(self):
        """Load dá»¯ liá»‡u hiá»‡n cÃ³ tá»« file JSON"""
        try:
            with open(TARGET_JSON, "r", encoding="utf-8") as f:
                self.existing_data = json.load(f)
                self.total_samples = len(self.existing_data)
                print(f"ğŸ“‚ ÄÃ£ load {self.total_samples} máº«u tá»« file hiá»‡n cÃ³")
        except (FileNotFoundError, json.JSONDecodeError):
            self.existing_data = []
            print("ğŸ†• Táº¡o file dataset má»›i")

    def generate_data(self, batches_per_topic=5):
        """HÃ m chÃ­nh Ä‘á»ƒ sinh dá»¯ liá»‡u"""
        try:
            while True:
                clear_output(wait=True)
                self._print_status()
                
                topic = TOPICS[self.current_topic_index]
                batch_data = self._process_batch(topic)
                
                if batch_data:
                    self._save_batch(batch_data)
                    self._update_state(batches_per_topic)
                
                time.sleep(1)  # TrÃ¡nh rate limit
                
        except KeyboardInterrupt:
            print("\nâ¹ Dá»«ng sinh dá»¯ liá»‡u")
            self._print_final_report()

    def _process_batch(self, topic: str) -> List[Dict]:
        """Xá»­ lÃ½ má»™t batch dá»¯ liá»‡u"""
        try:
            prompt = self._generate_prompt(topic)
            print(f"\nğŸ“ Prompt: {prompt[:150]}...")
            
            response = client.chat.completions.create(
                model="deepseek-chat",
                messages=self._prepare_messages(prompt),
                temperature=0.85,
                max_tokens=1024,
                response_format={"type": "json_object"}
            )
            
            reply = response.choices[0].message.content
            print(f"ğŸ¤– Pháº£n há»“i: {reply[:100]}...")
            
            batch_data = json.loads(reply).get("data", [])
            print(f"âœ¨ Sinh thÃ nh cÃ´ng {len(batch_data)} máº«u")
            return batch_data
            
        except Exception as e:
            print(f"âŒ Lá»—i khi xá»­ lÃ½ batch: {str(e)}")
            return []

    def _prepare_messages(self, prompt: str) -> List[Dict]:
        """Chuáº©n bá»‹ messages cho API"""
        messages = []
        if self.current_topic_call_count > 0:
            messages.extend(self.conversation_history)
        messages.append({"role": "user", "content": prompt})
        return messages

    def _generate_prompt(self, topic: str) -> str:
        """Táº¡o prompt phÃ¹ há»£p"""
        if self.current_topic_call_count == 0:
            return self._full_prompt(topic)
        return self._short_prompt(topic)

    def _full_prompt(self, topic: str) -> str:
        return FULL_PROMPT_TEMPLATE

    def _short_prompt(self, topic: str) -> str:
        return SHORT_PROMPT_TEMPLATE

    def _save_batch(self, batch_data: List[Dict]):
        """LÆ°u batch dá»¯ liá»‡u vÃ o file"""
        self.existing_data.extend(batch_data)
        self.total_samples += len(batch_data)
        
        with open(TARGET_JSON, "w", encoding="utf-8") as f:
            json.dump(self.existing_data, f, ensure_ascii=False, indent=2)

    def _update_state(self, max_calls: int):
        """Cáº­p nháº­t tráº¡ng thÃ¡i generator"""
        self.current_topic_call_count += 1
        
        if self.current_topic_call_count >= max_calls:
            self.current_topic_index = (self.current_topic_index + 1) % len(TOPICS)
            self.current_topic_call_count = 0
            self.conversation_history = []
            print(f"\nğŸ”„ Chuyá»ƒn sang topic: {TOPICS[self.current_topic_index]}")

    def _print_status(self):
        """Hiá»ƒn thá»‹ tráº¡ng thÃ¡i hiá»‡n táº¡i"""
        print(f"""
=== WAIFU DATASET GENERATOR ===
ğŸ“Š Tá»•ng máº«u: {self.total_samples}
ğŸ“Œ Topic hiá»‡n táº¡i: {TOPICS[self.current_topic_index]} 
ğŸ”„ Sá»‘ batch: {self.current_topic_call_count + 1}/5
=============================
        """)

    def _print_final_report(self):
        """BÃ¡o cÃ¡o cuá»‘i cÃ¹ng"""
        print(f"""
=== Káº¾T QUáº¢ CUá»I CÃ™NG ===
ğŸ“‚ File: {TARGET_JSON}
ğŸ“Š Tá»•ng máº«u: {self.total_samples}
ğŸ¯ Topics Ä‘Ã£ xá»­ lÃ½: {len(TOPICS)}
========================
        """)
```
